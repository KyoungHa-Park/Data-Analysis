```python
import pandas as pd
import os
from os import listdir
from datetime import datetime, timedelta

import numpy as np


# 그래프
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
path = 'C:/windows/Fonts/malgun.ttf'                           # 폰트:맑은고딕
font_name = fm.FontProperties(fname=path, size=24).get_name()  
plt.rc('font', family=font_name)
```

```python
##########
# 자료 불러오기
#####

# 경로 설정
directory_name = "./data_naver"

# 파일명 가져오기
file_names = listdir(directory_name)
# file_names = file_names[1:]

# file_names = file_names[1:]
```

```python
file_names[0][-15:-5]
```

```python
data = pd.DataFrame()

# 전체 자료 불러오기 & 병합
for open_file in file_names:
    df_temp = pd.read_excel(directory_name +'/'+ open_file, 0)
    df_temp['DATE'] = open_file[-15:-5]
    data = pd.concat([data, df_temp], ignore_index=True)
```

```python
data.tail(2)
```

```python
# pd.read_excel(directory_name +'/'+ open_file, 0)
```

```python
data['DATE'] = pd.to_datetime(data['DATE'],  format = '%Y-%m-%d')
```

```python
s_dt1= (datetime.today()- timedelta(0)).strftime('%Y%m%d')
excel_name = s_dt1+ '_상품_인구통계_v1.xlsx'
writer = pd.ExcelWriter(excel_name)

data.to_excel(writer, 'Sheet1')

writer.save()
writer.close()
```
