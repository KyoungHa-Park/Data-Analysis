```python
# 빠른 인코딩
lb_make = LabelEncoder()
df1["AGE_GUBUN"] = df1["AGE_GUBUN"].astype('category').cat.codes
df1['GENDER'] = df1['GENDER'].astype('category').cat.codes
df1['FAMILY'] = df1['FAMILY'].astype('category').cat.codes
df1['체험팩구매'] = df1['체험팩구매'].astype('category').cat.codes
df1['EVENT_before_BUY'] = df1['EVENT_before_BUY'].astype('category').cat.codes
df1['EVENT_AFTER_BUY'] = df1['EVENT_AFTER_BUY'].astype('category').cat.codes
df1['EVENT_DEAL_BUY'] =  df1['EVENT_DEAL_BUY'].astype('category').cat.codes
```

```python
df1_A = df1[df1['EVENT_DEAL_BUY']==1]
df1_B = df1[df1['EVENT_DEAL_BUY']==0].sample(2000)
```

```python
df1_fianl = pd.concat([df1_A, df1_B], ignore_index=True)
```

```python
# 자료 전처리

feature1= [
#     'AGE',
           'MILK_CHANGER', 'DIETER',    'ORGANIC_PREMIUM', 'MEAL_REPLACEMENT', 
           'HOME_COOK',    'WELL_AGING','TASTE_SEEKER',    'OH_MY_BABY', 
#            'PROMOTION_SENSITIVE', 'caffeine_SENSITIVE', 'SNS_USER','FIRST_TAKER'
           ]

data_X = df1_fianl[feature1].copy()
data_Y = df1_fianl["EVENT_DEAL_BUY"].copy()

# data split(train / test)
X1_train, X1_test, y1_train, y1_test = train_test_split(data_X, data_Y, test_size=0.2, random_state=0)

```

```python
X1_train.info()
```

```python
# Compare Algorithms
import pandas
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

from sklearn.neural_network import MLPClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
# from sklearn.gaussian_process.kernels import RBF

# load dataset : array type
X = data_X.values
Y = data_Y.values

# prepare configuration for cross validation test harness
seed = 7

# prepare models
models = []
models.append(('CART', DecisionTreeClassifier()))
models.append(('RF', RandomForestClassifier()))
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
models.append(('MLP', MLPClassifier()))
models.append(('GP', GaussianProcessClassifier()))
# models.append(('RBF', RBF()))

# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())

# boxplot algorithm comparison
mpl.rc('figure', figsize=(12, 6))

fig = plt.figure()
# plt.axvspan(0.6, 2.4, color='gray', alpha=0.5)
fig.suptitle('모델링 결과 비교')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
#check the optimal k value
from sklearn.cluster import KMeans
ks = range(1, 15)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(df1_fianl)
    inertias.append(model.inertia_)

plt.figure(figsize=(10,5))
plt.style.use('bmh')
plt.plot(ks, inertias, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(ks)
plt.show()
```

```python
# 모형 실행 : Decision Tree
model_tree2 = DecisionTreeClassifier(criterion='entropy', max_depth=2).fit(X1_train, y1_train)

# 모형 시각화 : Decision Tree
command_buf1 = io.StringIO()
export_graphviz(model_tree2, 
                out_file=command_buf1, 
                feature_names=feature1)
graph1 = pydot.graph_from_dot_data(command_buf1.getvalue())[0]
image1 = graph1.create_png()
Image(image1)
```

```python
# 모형 실행 : Decision Tree
model_tree2 = DecisionTreeClassifier(criterion='entropy', max_depth=2).fit(df1[feature1], df1["EVENT_DEAL_BUY"])

# 모형 시각화 : Decision Tree
command_buf1 = io.StringIO()
export_graphviz(model_tree2, 
                out_file=command_buf1, 
                feature_names=feature1)
graph1 = pydot.graph_from_dot_data(command_buf1.getvalue())[0]
image1 = graph1.create_png()
Image(image1)
```

```python
# 모형 실행 : Decision Tree
model_tree2 = DecisionTreeClassifier(criterion='entropy', max_depth=2).fit(X1_train, y1_train)

# 모형 중요변수 : Decision Tree
df2_1_0 = pd.DataFrame(feature1, columns =['feature'])
df2_1_1 = pd.DataFrame(model_tree2.feature_importances_, columns =['importances'])
model1_1 = pd.concat([df2_1_0, df2_1_1], axis=1)

# 모형 시각화 : 중요 변수 화인
mpl.rc('figure', figsize=(6, 5))
sns.barplot(y= model1_1['feature'], x= model1_1['importances'], color ='b')
sns.despine(left=True, bottom=True)
plt.title('Decision Tree Model',fontsize=15)

plt.show()
```
