```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
import matplotlib as mpl
%matplotlib inline

import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
import warnings
warnings.filterwarnings('ignore')
import io
import graphviz 
from sklearn.tree import export_graphviz
import pydot
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from IPython.core.display import Image
```

```python
df = pd.read_excel('./200723_소잘_설문응답_딜구매추가_v2.xlsx')
```

```python
sns.heatmap(df.isnull(), cbar=False)
```

```python
df['EVENT_AFTER_BUY'] = df['EVENT_AFTER_BUY'].fillna('N')
```

```python
sns.heatmap(df.isnull(), cbar=False)
```

```python
df.columns
```

```python
del df['UNFY_MMB_NO']
```

```python
df = df.dropna()
```

```python
df1 = df[['AGE', 'AGE_GUBUN', 'GENDER', 'FAMILY', '체험팩구매',
       'EVENT_before_BUY', 'EVENT_AFTER_BUY', 'EVENT_DEAL_BUY',
       'CATEGORY_COUNT', 'MILK_CHANGER', 'DIETER', 'ORGANIC_PREMIUM',
       'MEAL_REPLACEMENT', 'HOME_COOK', 'WELL_AGING', 'TASTE_SEEKER',
       'OH_MY_BABY', 'PROMOTION_SENSITIVE', 'caffeine_SENSITIVE', 'SNS_USER',
       'FIRST_TAKER']]
```

```python
# 빠른 인코딩
lb_make = LabelEncoder()
df1["AGE_GUBUN"] = df1["AGE_GUBUN"].astype('category').cat.codes
df1['GENDER'] = df1['GENDER'].astype('category').cat.codes
df1['FAMILY'] = df1['FAMILY'].astype('category').cat.codes
df1['체험팩구매'] = df1['체험팩구매'].astype('category').cat.codes
df1['EVENT_before_BUY'] = df1['EVENT_before_BUY'].astype('category').cat.codes
df1['EVENT_AFTER_BUY'] = df1['EVENT_AFTER_BUY'].astype('category').cat.codes
df1['EVENT_DEAL_BUY'] =  df1['EVENT_DEAL_BUY'].astype('category').cat.codes
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
df1_PCA = pca.fit_transform(df1[['AGE', 'AGE_GUBUN', 'GENDER', 'FAMILY', '체험팩구매',
       'EVENT_before_BUY', 'EVENT_AFTER_BUY', 
       'CATEGORY_COUNT', 'MILK_CHANGER', 'DIETER', 'ORGANIC_PREMIUM',
       'MEAL_REPLACEMENT', 'HOME_COOK', 'WELL_AGING', 'TASTE_SEEKER',
       'OH_MY_BABY', 'PROMOTION_SENSITIVE', 'caffeine_SENSITIVE', 'SNS_USER',
       'FIRST_TAKER']])

df1_PCA =pd.DataFrame(df1_PCA) 
```

```python
df1_PCA.info()
```

```python
# 자료 전처리

feature1= [0,1,2]

data_X = df1_PCA[feature1].copy()
data_Y = df1["EVENT_DEAL_BUY"].copy()

# data split(train / test)
X1_train, X1_test, y1_train, y1_test = train_test_split(data_X, data_Y, test_size=0.8, random_state=0)

```

```python
# Compare Algorithms
import pandas
import matplotlib.pyplot as plt
from sklearn import model_selection
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier

from sklearn.neural_network import MLPClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
# from sklearn.gaussian_process.kernels import RBF

# load dataset : array type
X = data_X.values
Y = data_Y.values

# prepare configuration for cross validation test harness
seed = 7

# prepare models
models = []
models.append(('CART', DecisionTreeClassifier()))
models.append(('RF', RandomForestClassifier()))
models.append(('LR', LogisticRegression()))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
models.append(('MLP', MLPClassifier()))
models.append(('GP', GaussianProcessClassifier()))
# models.append(('RBF', RBF()))

# evaluate each model in turn
results = []
names = []
scoring = 'accuracy'
for name, model in models:
    kfold = model_selection.KFold(n_splits=10, random_state=seed)
    cv_results = model_selection.cross_val_score(model, X, Y, cv=kfold, scoring=scoring)
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())

# boxplot algorithm comparison
mpl.rc('figure', figsize=(12, 6))

fig = plt.figure()
plt.axvspan(0.9, 1.0, color='gray', alpha=0.5)
fig.suptitle('모델링 결과 비교')
ax = fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```

```python
#check the optimal k value
from sklearn.cluster import KMeans
ks = range(1, 15)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(df1_PCA)
    inertias.append(model.inertia_)

plt.figure(figsize=(10,5))
plt.style.use('bmh')
plt.plot(ks, inertias, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(ks)
plt.show()
```

```python
#  max_depth=2 기준
```

```python
%%time 

# data 준비
data = df1_PCA

# model 생성 
model = KMeans(n_clusters=2,random_state=0)

# 원자료 + prediction
# feature = data[[0,1,2,3,4,5]]
feature = data[[0,1,2]]
predict = pd.DataFrame(model.fit_predict(feature))
predict.columns=['predict2']
r = pd.concat([feature,predict],axis=1)

# 그래프 출력
sns.pairplot(r, hue="predict2", height = 2 )
plt.show()
```

```python
# 모형 실행 : Decision Tree
model_tree1 = DecisionTreeClassifier(criterion='entropy', max_depth=2, min_samples_leaf=10).fit(X1_train, y1_train)

# 모형 시각화 : Decision Tree
command_buf1 = io.StringIO()
export_graphviz(model_tree1, 
                out_file=command_buf1, 
                feature_names=feature1)
graph1 = pydot.graph_from_dot_data(command_buf1.getvalue())[0]
image1 = graph1.create_png()
Image(image1 )
```

```python
# 모형 실행 : Decision Tree
model_tree1 = DecisionTreeClassifier(criterion='entropy', max_depth=2, min_samples_leaf=10).fit(X1_train, y1_train)

# 모형 중요변수 : Decision Tree
df2_1_0 = pd.DataFrame(feature1, columns =['feature'])
df2_1_1 = pd.DataFrame(model_tree1.feature_importances_, columns =['importances'])
model1_1 = pd.concat([df2_1_0, df2_1_1], axis=1)

# 모형 시각화 : 중요 변수 화인
mpl.rc('figure', figsize=(6, 5))
sns.barplot(y= model1_1['feature'], x= model1_1['importances'], color ='b')
sns.despine(left=True, bottom=True)
plt.title('Decision Tree Model',fontsize=15)

plt.show()
```
