### 변경 전

```python
# 결과 시각화
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('./설문_sample.xlsx')
```

```python
df.head(2)
```

```python
mpl.rc('figure', figsize=(15, 6))

plt.subplot(221)
sns.heatmap(df.isnull(), cbar=False)

plt.subplot(222)
plt.bar(range(len(df)),df["FAMILY"])

plt.subplot(224)
sns.barplot(data=df, x="FAMILY", y="AGE")

plt.show()
```
