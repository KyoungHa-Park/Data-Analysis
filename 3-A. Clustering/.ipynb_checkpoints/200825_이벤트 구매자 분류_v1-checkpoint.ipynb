```python
# data_load : version 2 

import cx_Oracle as oci
import pandas as pd

# 접속정보
con_text = open("C:/conn_CRM.txt", 'r', encoding='utf8')
con_text = con_text.read()

# DB 접속
conn = oci.connect(con_text)

# query 열기
f = open("./query_v1.txt", 'r')
line = f.read()
f.close()

## query 실행( =자료 추출)
df = pd.read_sql(line, con = conn)
```

### 2.EDA

```python
#누락 자료 확인
sns.heatmap(df.isnull(), cbar=False)
plt.show()
```

```python
#주요 변수 확인 : 연령대
plt.figure(figsize=(6,4))
plt.plot(df.groupby(df['AGE_GUBUN'])['AGE_GUBUN'].count())
plt.xticks(rotation=270)
plt.show()
```

```python
#주요 변수 확인 : 가입경로
plt.figure(figsize=(12,4))
plt.plot(df.groupby(df['NTRY_PATH'])['REG_DTM_GUBUN'].count())
plt.xticks(rotation=270)
plt.show()
```

```python
#주요 변수 확인 : 가입유형
plt.pie(df.groupby(df['REG_DTM_GUBUN'])['REG_DTM_GUBUN'].count(), startangle=90)
plt.show()

df.groupby(df['REG_DTM_GUBUN'])['REG_DTM_GUBUN'].count()
```

```python
#주요 변수 확인 : 성별
plt.pie(df.groupby(df['GENDER'])['GENDER'].count(), startangle=90)
plt.show()
df.groupby(df['GENDER'])['GENDER'].count()
```
