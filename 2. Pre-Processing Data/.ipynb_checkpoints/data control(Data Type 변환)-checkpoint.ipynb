## 3.value type 변환

```python
df1.head(1)
```

```python
df2 = df1.copy()
df2['REG_DATE']  = '2019-01-01'
df2['REG_DATE1'] = pd.to_datetime(df2['REG_DATE'],format=None) # 2-2. object to datetime64
df2['REG_DATE2'] = df2['REG_DATE'].astype(object)               # 2-1. datetime64 to object 
df2['YYYYMM']    = df2['REG_DATE'].str[:-3]                     # object to slice 

df2['class'] = pd.Categorical(df2['class'])   # 2-3. object to Categorical
df2['class'] = df2['class'].cat.codes         # 2-3. Categorical code reputation

# age = feature_column.numeric_column("age")
```
