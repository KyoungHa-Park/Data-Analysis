```python
import networkx as nx
from pyvis.network import Network

def df_to_network2(open_file, keyword1, keyword2):
    df = pd.read_excel('./'+ open_file,'Sheet1')
    
    mpl.rc('figure', figsize=(8, 6), facecolor='#eeeeee' )
    plt.xticks([]); plt.yticks([])
    G = nx.from_pandas_edgelist(df, 'from', 'to',  edge_attr=["weight"], create_using=nx.Graph())
    color_map = []
    node_size = []
    for node in G:
        if node in keyword1:
            color_map.append('#33A8FF')
            node_size.append(1200)
        elif node in keyword2:
            color_map.append('#33A80F')
            node_size.append(1200)
        else:
            color_map.append('#c8cbcc')
            node_size.append(300)
    # node_size2 = []
    # for i in range(0, len(node_size)):
    #     node_size2.append(node_size[i] * df["weight"][i])
    nx.draw_networkx(G, with_labels=True, edgecolors='gray', font_family=font_name, font_size=10,
                     node_shape ='o', node_size =  node_size, node_color = color_map)
    plt.title('Pyhton 활용 현황', fontsize='x-large')
    plt.show()
    return
```

```python
keyword_database = ['SQL\n(Oracle)', 'SQL\n(MySQL)', 'SQL\n(BigQuery)', 'Data\n(Excel)']
keyword_solution = ['API\n(Vision)', 'API\n(Bard)',  'Google\nStorage', 'Google\nFunction', 'SMTP\n(gmail)']

df_to_network2('workList_by_NetworkGraph.xlsx', keyword_database, keyword_solution)
```

```python
import networkx as nx
from pyvis.network import Network



def df_to_network1(dataset):
    mpl.rc('figure', figsize=(8, 8), facecolor='#eeeeee' )
    G = nx.from_pandas_edgelist(dataset, 'from', 'to',  edge_attr=["weight"], create_using=nx.Graph())
    nx.draw_networkx(G, pos = nx.shell_layout(G), node_color='#c8cbcc', with_labels=True, edgecolors='gray', font_family=font_name,
                     node_shape ='o', width = list(dataset['weight']*0.2) )
    plt.title('키워드의 연관단어', fontsize='x-large')
    return
```

```python
open_file = 'workList_by_NetworkGraph.xlsx'
df = pd.read_excel('./'+ open_file,'Sheet1')

df_to_network1(df)
```
