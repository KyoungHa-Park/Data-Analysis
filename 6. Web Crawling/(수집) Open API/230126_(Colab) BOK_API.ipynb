```python
from datetime import datetime, timedelta
import pandas as pd
import requests
from google.cloud import bigquery
import pandas_gbq
from dateutil.relativedelta import relativedelta


# ì €ì¥ ìœ„ì¹˜ ì„¤ì •
project = 'boxwood-**************'
client = bigquery.Client(project)
dataset_id = 'openAPI'
table_id = 'BOK_StatisticSearch'
table_name = dataset_id +'.' +table_id


def API_data_get(STAT_NAME, ITEM_CODE1, ITEM_CODE2):
    api_ket = '__api_key__'
    # start_month = (datetime.today() - relativedelta(months=0)).strftime('%Y%m')
    # end_month   = (datetime.today() - relativedelta(months=0)).strftime('%Y%m')
    start_month = '201901'
    end_month   = '202003'
    url = 'https://ecos.bok.or.kr/api/StatisticSearch/'+api_ket+'/json/kr/1/50/'+STAT_NAME+'/M/'+start_month+'/'+end_month+'/'+ITEM_CODE1+'/'+ITEM_CODE2+'/?/?'
    response = requests.get(url)

    df_temp = pd.DataFrame(response.json()['StatisticSearch']['row'])
    df_temp['TIME'] = pd.to_datetime(df_temp['TIME'], format='%Y%m')
    df_temp['DATA_VALUE'] = df_temp['DATA_VALUE'].astype(float)
    df_temp = df_temp.fillna(value='')
    return df_temp

def get_data(t1):
    get_list = [['513Y001', 'E1000', '?'     ],
                ['521Y001', 'A001' , '?'     ],
                ['511Y002', 'FMCCC', '99988' ],] 
    dataset = pd.DataFrame()
    for STAT_NAME_, ITEM_CODE1_, ITEM_CODE2_ in get_list:
        rows_to_insert = API_data_get(STAT_NAME_, ITEM_CODE1_, ITEM_CODE2_)
        dataset = pd.concat([dataset, rows_to_insert ], ignore_index = True )
    # pandas_gbq.to_gbq(dataset, table_name, project_id=project, if_exists = 'append' )
    # return u'Hellow word'
    return dataset
```
