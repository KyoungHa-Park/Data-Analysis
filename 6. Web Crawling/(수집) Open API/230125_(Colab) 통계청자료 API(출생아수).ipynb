```python
import pandas as pd
import requests
from google.cloud import bigquery
import pandas_gbq

# 저장 위치 설정

project = 'boxwood-**************'
client = bigquery.Client(project)
dataset_id = 'openAPI'
table_id = 'kosis_Birthrate'
table_name = dataset_id +'.' +table_id

token = '__token_values__'
# code = ['1364', '']

def get_data(t1):
    url = 'http://kosis.kr/kosisapi/service/IndicatorService/IndIdDetailSearchRequest?STAT_JIPYO_ID=1364&SRV_RN=1&SRV_RN=1&serviceKey='+token
    response = requests.get(url)
    dataset = pd.read_xml(response.text, xpath="//item")
    dataset['val'] = dataset['val'].str.replace(',', '')
    dataset['val'] = dataset['val'].astype(int)
    dataset['prdDe'] = pd.to_datetime(dataset['prdDe'], format='%Y%m')     
    dataset['statJipyoId'] = dataset['statJipyoId'].astype(int)
    # pandas_gbq.to_gbq(dataset, table_name, project_id=project, if_exists = 'append' )
    return dataset
    # return u'Hellow word'
```
