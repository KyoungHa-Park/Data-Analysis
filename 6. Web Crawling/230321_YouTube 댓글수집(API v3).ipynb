```python
import pandas
from pandas import json_normalize
from googleapiclient.discovery import build
from jinja2 import Template
```

```python
DEVELOPER_KEY='__DEVELOPER_KEY__'

link = 'https://youtu.be/CVOhpOAd_Lw'
video_id = 'CVOhpOAd_Lw'
```

```python
comments = list()
api_obj = build('youtube', 'v3', developerKey=DEVELOPER_KEY)
response = api_obj.commentThreads().list(part='snippet,replies', videoId=video_id, maxResults=100).execute()
```

```python
df_temp = json_normalize(response['items'])

df = df_temp[['snippet.topLevelComment.snippet.authorDisplayName', 'snippet.topLevelComment.snippet.textDisplay', 'snippet.topLevelComment.snippet.publishedAt', 'snippet.topLevelComment.snippet.likeCount', 'snippet.topLevelComment.snippet.authorChannelUrl' ]]
col_nm ={'snippet.topLevelComment.snippet.authorDisplayName': "authorDisplayName",
         'snippet.topLevelComment.snippet.textDisplay' : 'textDisplay',
         'snippet.topLevelComment.snippet.publishedAt' : 'publishedAt', 
         'snippet.topLevelComment.snippet.authorChannelUrl' : 'authorChannelUrl',
         'snippet.topLevelComment.snippet.likeCount' : 'likeCount' }

df = df.rename(columns=col_nm)
df['publishedAt'] = pd.to_datetime(df['publishedAt'])
df['publishedAt']= df['publishedAt'].dt.tz_convert('Asia/Seoul')
```
