```python

import cx_Oracle as oci
import pandas as pd


from sklearn.datasets import load_iris
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot  as plt
import seaborn as sns

```

```python
# version 2 : test1

# 접속정보
con_text = open("C:/Users/MAEIL/conn_info_MMBSHIP.txt", 'r', encoding='utf8')
con_text = con_text.read()

# DB 접속
conn = oci.connect(con_text)

## query 실행( =자료 추출)
df = pd.read_sql("select * from temp_pkh_selmun2" , con = conn)
```

```python
df.columns
```

```python
df1 = df[['UNFY_MMB_NO','MILK_CHANGER', 'DIETER',
       'ORGANIC_AND_PREMIUM', 'MEAL_REPLACEMENT', 'HOME_COOK',
       'PROMOTION_SENSITIVE', 'WELL_AGING', 'TASTE_SEEKER',
       'SENSITIVE_TO_CAFFINE', 'SNS_USER', 'OH_MY_BABY', 'FIRST_TAKER']]
```

```python
df1 = df1.dropna()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
df1_PCA = pca.fit_transform(df1[['MILK_CHANGER', 'DIETER','ORGANIC_AND_PREMIUM', 
                                 'MEAL_REPLACEMENT', 'HOME_COOK','PROMOTION_SENSITIVE', 
                                 'WELL_AGING', 'TASTE_SEEKER','SENSITIVE_TO_CAFFINE', 
                                 'SNS_USER','OH_MY_BABY', 'FIRST_TAKER']])

df1_PCA =pd.DataFrame(df1_PCA) 
```

```python
df1_PCA.head()
```

```python
#check the optimal k value
ks = range(1, 15)
inertias = []

for k in ks:
    model = KMeans(n_clusters=k)
    model.fit(df1_PCA)
    inertias.append(model.inertia_)

plt.figure(figsize=(10,5))
plt.style.use('bmh')
plt.plot(ks, inertias, '-o')
plt.xlabel('Number of clusters, k')
plt.ylabel('Inertia')
plt.xticks(ks)
plt.show()
```

```python
%time 

# data 준비
data = df1_PCA

# model 생성 
model = KMeans(n_clusters=2,random_state=0)

# 원자료 + prediction
# feature = data[[0,1,2,3,4,5]]
feature = data[[0,1]]
predict = pd.DataFrame(model.fit_predict(feature))
predict.columns=['predict5']
r = pd.concat([feature,predict],axis=1)

# 그래프 출력
sns.pairplot(r, hue="predict5", height = 2 )
plt.show()
```
