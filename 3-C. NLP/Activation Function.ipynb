### Activation Function
+ [x] 시각화
+ [ ] 결과비교

```python
import tensorflow as tf
import numpy as np
import matplotlib.pylab as plt
```

### 시각화

```python
def activation_function_vis(plot_location, identifier ):
    x=  np.arange(-5.0, 5.0, 0.1)
    y = tf.keras.activations.get(identifier)(x)
    plt.subplot(2,5, plot_location )
    plt.title(identifier)
    return plt.plot(x, y) 
```

```python
#그래프 그러보기
plt.figure(figsize=(15, 5))

activation_function_vis(1, 'sigmoid')
activation_function_vis(2, 'relu')
activation_function_vis(3, 'tanh')
activation_function_vis(4, 'swish')
activation_function_vis(5, 'softsign')

activation_function_vis(6, 'softplus')
activation_function_vis(7, 'selu')
activation_function_vis(8, 'softsign')
activation_function_vis(9, 'gelu')
activation_function_vis(10, 'elu')

plt.show()
```
