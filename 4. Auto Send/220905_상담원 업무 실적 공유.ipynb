```python

###############
# 내용공유(e-mail)
#####


# 이메일 양식(내용) frame 구성하기
message_tmpl="""
<html>
    <!DOCTYPE html>
    <html>
    <head lang ="ko">
      <meta charset ="UTF-8">
      <title>{{title1}}</title>
    </head>
    <style type="text/css">
        table, th, td {
            font-size: 8pt;
            border-collapse: collapse;
            font-family: Malgun Gothic;
            white-space: nowrap;
            border: 1px solid #bcbcbc;
            }
        thead tr {
            background-color: #0d47a1;
            color: #ffffff;
            }
    </STYLE>
    <BODY style="FONT-SIZE: 10pt; font-family: Malgun Gothic">
    <br>
    안녕하세요,<br>
    <br>
    0to7 이슈 관련, CS 응대내용 공유 드립니다.<br>
    (최종 수집 : {{d0}}시) <br>
    <br>
    <p>{{ df2_html }}</p>
    <br>
    감사합니다.<br>
    <br>
    </BODY>
</html>
"""

render_data = {
    'd0' : (datetime.today()).strftime('%m/%d, %H'),
    'df2_html' : df2[:20].to_html(),
}

t= Template(message_tmpl)
html_report = t.render(render_data)
```

```python

# 발송 변수 설정

# 받는사람 : mulit 발송
to_users = [ '**********@maeil.com' ]


# 받는사람 : test 발송
# to_users = ['**********@maeil.com']

# 메일제목
now = datetime.today()    # 현재
date = now.strftime('%Y%m%d')
subject = "[TEMP] "+date[2:] + "_CS 응대현황(발송전용)"

# 본문
html = html_report

# 첨부파일 추가
# attach = './' + excel_name

# 이메일 발송
se.send_email(to_users, subject, html, None, None)

# 생성된 파일 삭제
# os.remove(excel_name)
# for graph_ in graph:
#     os.remove('./' + graph_)
```
